```python
import numpy as np

np.random.seed(100)

x = np.random.normal(0, 10, 100)
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x) # Use hist() function to make histogram, x is the input data
plt.show() # Display the plot
```

```python
plt.hist(x, color='green')
plt.show()
```

```python
plt.hist(x, color='skyblue')
plt.show()
```


## Bin size
The bin input argument allow us to specify how many bins or bars you see in the plot. The total number of bins (bars) is equal to the number that we assign to the bin input argument. A bin represents the interval (or literally the range of values specified at the bin (bar).

At default we can see a bin size of *10.

```python
plt.hist(x)
plt.show()
```

Now, we will set the bin size to 5 and so there will be 5 bars.

```python
plt.hist(x, bins=5)
plt.show()
```

### Title and axis labels

```python

fig, ax = plt.subplots()

ax.hist(x)
ax.set_title('Histogram plot') # Title
ax.set_xlabel('X axis label')  # X axis label
ax.set_ylabel('Y axis label')  # Y axis label

plt.show()
```

### Relative width of bars
The relative width or rwidth cn be a value from 0 to 1.

```python
plt.hist(x, rwidth=0.1)
plt.show()
```

```python
plt.hist(x, rwidth=0.5)
plt.show()
```

```python
plt.hist(x, rwidth=0.9)
plt.show()
```

### Orientation
By default, the orientation of bars in the plot is vertical. Here, we will try setting the value to 'horizontal'.

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x, orientation='horizontal')
plt.show()
```

### Type of histogram bar
By default, we have the bar, here we will try using the step value.

```python
plt.hist(x, histtype='step') # 'bar', 'barstacked', 'step', 'stepfilled'
plt.show()
```

### Edge color of bar
The edge color of bars can be specified by the ec input argument.

```python
plt.hist(x, ec = 'black')
plt.show()
```
